```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib as mpl
import contextily as ctx
import pandas as pd
from matplotlib.colors import ListedColormap

# Cargar el archivo GeoJSON que contiene los datos geoespaciales
geojson_url = "C:/Users/Brandon.Lopez/Documents/Trabajo/2307 - Tijuana - Ensenada/Demanda/mapas/AGEBs Ensenada.geojson"   # Reemplaza con la ruta al archivo GeoJSON
gdf = gpd.read_file(geojson_url)

# Set the figure size using plt.figure()
plt.figure(figsize=(15, 15))

# Create the axes object and set the x-axis and y-axis limits
ax = plt.gca()
ax.set_xlim(-116.75, -116.5)  # Adjust the values as per your desired zoom level
ax.set_ylim(31.75, 31.95)      # Adjust the values as per your desired zoom level

# Add the basemap using 'ctx.add_basemap'
ctx.add_basemap(ax, crs=gdf.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)

# Plot the GeoDataFrame using 'gdf.plot()' and specify the column for color
color = gdf['Poblacion_total']  # Replace 'CSV_Poblacion_Total' with the appropriate column name from your GeoDataFrame
plot = gdf.plot(column=color, cmap='Reds', linewidth=0.8, edgecolor='0.2', legend=False, alpha=0.80, ax=ax)

# Create the colorbar as a legend
cbaxes = plt.gcf().add_axes([0.20, 0.20, 0.015, 0.20])
normlegend = mpl.colors.Normalize(vmin=color.min(), vmax=color.max())
legend = mpl.colorbar.ColorbarBase(cbaxes, norm=normlegend, cmap='Reds', orientation="vertical")
cbaxes.set_title('Población total', fontsize=20, color='black', fontname="Arial")
legend.ax.tick_params(labelcolor='black', size=10, labelsize=15,)


# Show the plot
plt.show()
```
